```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,3,4])
ypoints = np.array([3,8,2,6])

plt.plot(xpoints, ypoints,marker = 'o', ms = 12, mec= 'b', mfc = 'r')
plt.show()
```
